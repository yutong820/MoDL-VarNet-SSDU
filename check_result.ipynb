```python
def get_event_file_paths(base_dir, subdirs, event_file_name, data_type='train'):
   
    event_files = []
    for subdir in subdirs:
        dir_path = os.path.join(base_dir, subdir, data_type)
        for file in os.listdir(dir_path):
            if event_file_name in file:
                event_files.append(os.path.join(dir_path, file))
    return event_files

def read_tensorboard_data(log_file, tag):

    steps, values = [], []
    for e in tf.compat.v1.train.summary_iterator(log_file):
        for v in e.summary.value:
            if v.tag == tag:
                steps.append(e.step)
                values.append(v.simple_value)
    return steps, values
# R=2
base_dir_mdol_test_2 = './results/modl/v2_folder2'
subdirs_mdol_test_2 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_file_name = 'events.out.tfevents'  
event_files_modl_test_2 = get_event_file_paths(base_dir_mdol_test_2, subdirs_mdol_test_2, event_file_name, 'test')

base_dir_varnet_test_2 = './results/varnet/v2_folder2'
subdirs_varnet_test_2 = [f'base_varnet,k={i}' for i in range(1, 11)]  
event_files_varnet_test_2 = get_event_file_paths(base_dir_varnet_test_2, subdirs_varnet_test_2, event_file_name, 'test')

base_dir_ssdu_test_2 = './results/ssdu/v2_folder2'
subdirs_ssdu_test_2 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_2 = get_event_file_paths(base_dir_ssdu_test_2, subdirs_ssdu_test_2, event_file_name, 'test')


# R=4
base_dir_mdol_test_4 = './results/modl/v2'
subdirs_mdol_test_4 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_file_name = 'events.out.tfevents'  
event_files_modl_test_4 = get_event_file_paths(base_dir_mdol_test_4, subdirs_mdol_test_4, event_file_name, 'test')

base_dir_varnet_test_4 = './results/varnet/v2'
subdirs_varnet_test_4 = [f'base_varnet,k={i}' for i in range(1, 11)]  
event_files_varnet_test_4 = get_event_file_paths(base_dir_varnet_test_4, subdirs_varnet_test_4, event_file_name, 'test')

base_dir_ssdu_test_4 = './results/ssdu/v2'
subdirs_ssdu_test_4 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_4 = get_event_file_paths(base_dir_ssdu_test_4, subdirs_ssdu_test_4, event_file_name, 'test')


# R=6
base_dir_mdol_test_6 = './results/modl/v2_folder6'
subdirs_mdol_test_6 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_file_name = 'events.out.tfevents'  
event_files_modl_test_6 = get_event_file_paths(base_dir_mdol_test_6, subdirs_mdol_test_6, event_file_name, 'test')

base_dir_varnet_test_6 = './results/varnet/v2_folder6'
subdirs_varnet_test_6 = [f'base_varnet,k={i}' for i in range(1, 11)]  
event_files_varnet_test_6 = get_event_file_paths(base_dir_varnet_test_6, subdirs_varnet_test_6, event_file_name, 'test')

base_dir_ssdu_test_6 = './results/ssdu/v2_folder6'
subdirs_ssdu_test_6 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_6 = get_event_file_paths(base_dir_ssdu_test_6, subdirs_ssdu_test_6, event_file_name, 'test')

# modl layers = 3
base_dir_mdol_test_layer3 = './results/modl/layer3'
subdirs_mdol_test_layer3 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_file_name = 'events.out.tfevents'  
event_files_modl_test_layer3 = get_event_file_paths(base_dir_mdol_test_layer3, subdirs_mdol_test_layer3, event_file_name, 'test')

# modl layer3 = 5
base_dir_mdol_test_layer5 = './results/modl/v2'
subdirs_mdol_test_layer5 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_files_modl_test_layer5 = get_event_file_paths(base_dir_mdol_test_layer5, subdirs_mdol_test_layer5, event_file_name, 'test')

# modl layer3 = 7
base_dir_mdol_test_layer7 = './results/modl/layer7'
subdirs_mdol_test_layer7 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_files_modl_test_layer7 = get_event_file_paths(base_dir_mdol_test_layer7, subdirs_mdol_test_layer7, event_file_name, 'test')

# modl layer3 = 9
base_dir_mdol_test_layer9 = './results/modl/layer9'
subdirs_mdol_test_layer9 = [f'base_modl,k={i}' for i in range(1, 11)] 
event_files_modl_test_layer9 = get_event_file_paths(base_dir_mdol_test_layer9, subdirs_mdol_test_layer9, event_file_name, 'test')

# ssdu blocks = 10
base_dir_ssdu_test_blocks10 = './results/ssdu/block10'
subdirs_ssdu_test_blocks10 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_blocks10 = get_event_file_paths(base_dir_ssdu_test_blocks10, subdirs_ssdu_test_blocks10, event_file_name, 'test')

# ssdu blocks = 15
base_dir_ssdu_test_blocks15 = './results/ssdu/v2'
subdirs_ssdu_test_blocks15 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_blocks15 = get_event_file_paths(base_dir_ssdu_test_blocks15, subdirs_ssdu_test_blocks15, event_file_name, 'test')

# ssdu blocks = 20
base_dir_ssdu_test_blocks20 = './results/ssdu/block20'
subdirs_ssdu_test_blocks20 = [f'base_ssdu,k={i}' for i in range(1, 11)]  
event_files_ssdu_test_blocks20 = get_event_file_paths(base_dir_ssdu_test_blocks20, subdirs_ssdu_test_blocks20, event_file_name, 'test')

def plot_data(log_files, tag, k_values, subplot_index):
    ax = plt.subplot(2, 3, subplot_index)

    k_values_plot = []
    values_plot = []

    for log_file, k_value in zip(log_files, k_values):
       
        _, values = read_tensorboard_data(log_file, tag)
        if values:  
            k_values_plot.extend([k_value] * len(values))  
            values_plot.extend(values)
            print(values_plot)
```

```python
from skimage.metrics import (
    peak_signal_noise_ratio as PSNR,
    structural_similarity as SSIM,
)

def normalization(data):
    return data / data.max()

def calc_psnr(x, y):
    x = normalization(x)
    y = normalization(y)
    return round(PSNR(x, y, data_range=x.max() - x.min()),2)


def calc_ssim(x, y):
    x = normalization(x)
    y = normalization(y)
    return round(SSIM(x, y, data_range=x.max() - x.min()),2)

modl_2_psnr = calc_psnr(gt_2,modl_2)
modl_2_ssim = calc_ssim(gt_2, modl_2)
varnet_2_psnr = calc_psnr(gt_2,varnet_2)
varnet_2_ssim = calc_ssim(gt_2,varnet_2)
ssdu_2_psnr = calc_psnr(gt_2, ssdu_2)
ssdu_2_ssim = calc_ssim(gt_2, ssdu_2)

modl_4_psnr = calc_psnr(gt_4,modl_4)
modl_4_ssim = calc_ssim(gt_4, modl_4)
varnet_4_psnr = calc_psnr(gt_4,varnet_4)
varnet_4_ssim = calc_ssim(gt_4,varnet_4)
ssdu_4_psnr = calc_psnr(gt_4, ssdu_4)
ssdu_4_ssim = calc_ssim(gt_4, ssdu_4)

modl_6_psnr = calc_psnr(gt_6,modl_6)
modl_6_ssim = calc_ssim(gt_6, modl_6)
varnet_6_psnr = calc_psnr(gt_6,varnet_6)
varnet_6_ssim = calc_ssim(gt_6,varnet_6)
ssdu_6_psnr = calc_psnr(gt_6, ssdu_6)
ssdu_6_ssim = calc_ssim(gt_6, ssdu_6)

print(modl_2_psnr)
print(modl_2_ssim)

print(varnet_2_psnr)
print(varnet_2_ssim)

print(ssdu_2_psnr)
print(ssdu_2_ssim)

print(modl_4_psnr)
print(modl_4_ssim)

print(varnet_4_psnr)
print(varnet_4_ssim)

print(ssdu_4_psnr)
print(ssdu_4_ssim)

print(modl_6_psnr)
print(modl_6_ssim)

print(varnet_6_psnr)
print(varnet_6_ssim)

print(ssdu_6_psnr)
print(ssdu_6_ssim)
```

```python
from skimage.metrics import (
    peak_signal_noise_ratio as PSNR,
    structural_similarity as SSIM,
)

def normalization(data):
    return data / data.max()

def calc_psnr(x, y):
    x = normalization(x)
    y = normalization(y)
    return PSNR(x, y, data_range=x.max() - x.min())


def calc_ssim(x, y):
    x = normalization(x)
    y = normalization(y)
    return SSIM(x, y, data_range=x.max() - x.min())
gt_4 = load_gt(gt_4_path)
print(calc_ssim(gt_4, modl_4))
```
